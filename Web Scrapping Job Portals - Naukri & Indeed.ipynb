```python
#import required libaries
from datetime import datetime
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import requests
import time
import csv

# function to generate url, return two urls for two portals
def get_url(position, location):
    
    indeed = 'https://in.indeed.com/jobs?q={}&l={}'
    naukri = 'https://www.naukri.com/{}-jobs-in-{}'
    
    url_indeed = indeed.format(position, location)
    url_naukri = naukri.format(position,location)
    
    return url_indeed, url_naukri

# fn fetch and return attributes of a job as a record for Indeed portal
def get_indeed_record(indeed_card):
    
    indeed_url = 'https://in.indeed.com' + indeed_card.find('a',class_='jcs-JobTitle css-jspxzf eu4oa1w0').get('href')
    
    indeed_title = indeed_card.find('span').get('title')
    
    indeed_company = indeed_card.find('span','companyName').text.strip()

    indeed_location = indeed_card.find('div','companyLocation').text.strip()
    
    try:
        indeed_salary = indeed_card.find('div','attribute_snippet').text
    except AttributeError:
        indeed_salary = ''

    indeed_exp = ''
    
    indeed_post_date = indeed_card.find('span','date').text.strip()
    
    today = datetime.today().strftime('%Y-%m-%d')
    
    indeed_summary = indeed_card.find('div','job-snippet').text.strip()
    
    indeed_record = (indeed_title, indeed_company, indeed_location, indeed_exp, indeed_post_date, today, indeed_summary, indeed_salary, indeed_url,'Indeed')
    
    return indeed_record

# fn fetch and return attributes of a job as a record for Naukri portal
def get_naukri_record(naukri_card):
    
    naukri_url = naukri_card.find('a',class_='title fw500 ellipsis').get('href')
    
    naukri_title = naukri_card.find('a',class_='title fw500 ellipsis').get('title')
    
    naukri_company = naukri_card.find('a',class_='subTitle ellipsis fleft').text.strip()
    
    Loc = naukri_card.find('li',class_='fleft grey-text br2 placeHolderLi location')
    try:
        naukri_location = Loc.find('span',class_='ellipsis fleft fs12 lh16').text.strip()
    except AttributeError:
        naukri_location = ''

    Sal = naukri_card.find('li',class_='fleft grey-text br2 placeHolderLi salary')
    naukri_salary = Sal.find('span',class_='ellipsis fleft fs12 lh16').text.strip()

    Exp = naukri_card.find('li',class_='fleft grey-text br2 placeHolderLi experience')
    try:
        naukri_exp = Exp.find('span',class_='ellipsis fleft fs12 lh16').text.strip()
    except AttributeError:
        naukri_exp = ''
    
    post_date = naukri_card.find("div",["type br2 fleft grey","type br2 fleft green"])
    naukri_post_date = post_date.find('span',class_='fleft fw500').text.strip()
    
    today = datetime.today().strftime('%Y-%m-%d')
    
    naukri_summary = naukri_card.find('div',class_='job-description fs12 grey-text').text.strip()
    
    naukri_record = (naukri_title, naukri_company, naukri_location, naukri_exp, naukri_post_date, today, naukri_summary, naukri_salary, naukri_url,'Naukri')
    
    return naukri_record

# Main fn which calls above fns and writes the records into csv
def main(position, location):
    records = []
    url = []
    
    geturl = get_url(position,location)
    
    for _ in geturl:
        url.append(_)

    
    while True:
        
        indeed = webdriver.Chrome('C:\\Users\\Nikhil\\anaconda3\\chromedriver.exe')
        indeed.get(url[0])

        time.sleep(10)

        indeed_soup = BeautifulSoup(indeed.page_source,'html5lib')
        
        indeed.close()
        
        indeed_cards = indeed_soup.find_all('div',class_ = 'cardOutline')
    
        for indeed_card in indeed_cards:
            record = get_indeed_record(indeed_card)
            records.append(record)
            
        with open('all.csv','w',newline= '', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['Title', 'Company', 'Location', 'Experience', 'Post_Date', 'Today', 'Summary', 'Salary', 'Url','Portal'])
            writer.writerows(records)
        try:
            url[0] = 'https://in.indeed.com' + indeed_soup.find('a',{'aria-label':'Next'}).get('href')
        except AttributeError:
            break
    
    while True:
        naukri = webdriver.Chrome('C:\\Users\\Nikhil\\anaconda3\\chromedriver.exe')
        naukri.get(url[1])
        
        time.sleep(3)
        
        naukri_soup = BeautifulSoup(naukri.page_source,'html5lib')
        
        naukri.close()
        
        naukri_cards = naukri_soup.find_all('article',class_='jobTuple bgWhite br4 mb-8')

        for naukri_card in naukri_cards:
            record = get_naukri_record(naukri_card)
            records.append(record)
            
        with open('all.csv','w',newline= '', encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(['Title', 'Company', 'Location', 'Experience', 'Post_Date', 'Today', 'Summary', 'Salary', 'Url','Portal'])
            writer.writerows(records)
            
        try:
            url[1] = 'https://www.naukri.com' + naukri_soup.find('a',class_='fright fs14 btn-secondary br2').get('href')
        except AttributeError:
            break
```

```python
main('data analyst','hyderabad')
```
